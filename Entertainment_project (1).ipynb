```python
# to check duplicates
df.duplicated().sum()    # it check entire rows duplicates eg:reenter of same value
for col in df.columns:
    num_duplicates = df[col].duplicated().sum()
    print(f"Column '{col}' has {num_duplicates} duplicate values.")
    
duplicate_summary = pd.DataFrame({
    'Column': df.columns,
    'Duplicate_Count': [df[col].duplicated().sum() for col in df.columns]
})

duplicate_summary
```

```python
column_duplicates = {
    col: df[col][df[col].duplicated()].unique().tolist()
    for col in df.columns
    if df[col].duplicated().sum() > 0
}

pd.DataFrame([
    {"Column": k, "Duplicated_Values": v}
    for k, v in column_duplicates.items()
])
```

# 1. Top Performers by Career Earnings 

```python

# Top 10 by earnings
top_earners = df.nlargest(10, 'Total_Career_Earnings_Lakhs')  #top N largest values
#print(top_earners)
# Barplot
sns.barplot(data=top_earners, x='Total_Career_Earnings_Lakhs', y='Entertainer', hue='Gender')
plt.title("Top 10 Entertainers by Career Earnings")
plt.xlabel("Earnings (Lakhs)")
plt.ylabel("Entertainer")
plt.tight_layout()
plt.show()
```



##  **Problem 1: Top Performers by Career Earnings**

## **Conclusion:**
Top entertainers by earnings have consistently higher show counts and awards. 
This suggests that sustained activity and recognition strongly correlate with financial success in the industry.


# 2. Age vs Earnings Relationship 

```python
sns.scatterplot(data=df, x='Age_Entertain ', y='Total_Career_Earnings_Lakhs', hue='Gender')
plt.title("Age at Career Start vs Total Career Earnings")
plt.xlabel("Age at Breakthrough")
plt.ylabel("Total Career Earnings (Lakhs)")
plt.show()
```

##  **Problem 2: Age vs Earnings Relationship**

## **Conclusion:**
There is no strict linear relationship, but many high earners began their careers in their late teens or early twenties, 
indicating that early entry might contribute to longer, more profitable careers.

# 3. Gender-wise Career Analysis 

```python
gender_stats = df.groupby('Gender').agg({
    'Entertainer': 'count',
    'Active_Years': 'mean',
    'Total_Career_Earnings_Lakhs': 'mean',
    'Total_Awards': 'sum'
})
#.reset_index()

print(gender_stats)
# barplot
sns.barplot(data=gender_stats, x='Gender', y='Total_Career_Earnings_Lakhs')
plt.title("Average Career Earnings by Gender")
plt.ylabel("Avg Earnings (Lakhs)")
plt.show()
```


##  **Problem 3: Gender-wise Career Analysis**

## **Conclusion:**
While gender distribution may be balanced, male entertainers often show slightly higher average earnings and career duration, 
though female entertainers may receive a comparable number of awards, 
pointing to equality in recognition but potential disparity in longevity or compensation.

# 4. Impact of Early Awards on Career

```python
early_award_group = df.groupby('Awarded_Early').agg({
    'Entertainer': 'count',
    'Active_Years': 'mean',
    'Total_Career_Earnings_Lakhs': 'mean'
}).reset_index()

sns.barplot(data=early_award_group, x='Awarded_Early', y='Total_Career_Earnings_Lakhs')
plt.title("Earnings by Early Award Status")
plt.xlabel("Awarded Early")
plt.ylabel("Avg Earnings (Lakhs)")
plt.show()
```


## **Problem 4: Impact of Early Awards on Career**

## **Conclusion:**
Entertainers who received awards early in their careers tend to have longer active years and earn more on average.
Early recognition appears to boost long-term career prospects and earnings potential.

# 5. City-wise Talent Hub

```python
city_summary = df.groupby('City').agg({
    'Entertainer': 'count',
    'Total_Awards': 'sum',
    'Total_Career_Earnings_Lakhs': 'mean'
}).reset_index().sort_values('Entertainer', ascending=False)

sns.barplot(data=city_summary.head(10), x='Entertainer', y='City')
plt.title("Top Cities by Entertainer Count")
plt.xlabel("Number of Entertainers")
plt.ylabel("City")
plt.show()
```

##  **Problem 5: City-wise Talent Hub**

## **Conclusion:**
Certain cities consistently produce more high-earning entertainers. These cities also have higher total awards, 
suggesting strong industry infrastructure or talent pipelines in those regions.

# 6. Award Winners vs Non-Winners 

```python
award_stats = df.groupby('Received_Award').agg({
    'Entertainer': 'count',
    'Active_Years': 'mean',
    'Total_Career_Earnings_Lakhs': 'mean'
}).reset_index()

sns.barplot(data=award_stats, x='Received_Award', y='Total_Career_Earnings_Lakhs')
plt.title("Earnings: Award Winners vs Non-Winners")
plt.show()
```

##  **Problem 6: Award Winners vs Non-Winners**

## **Conclusion:**
Award winners generally have longer careers and significantly higher average earnings compared to non-winners.
Awards act as a catalyst for both visibility and financial success.

# 10. Earnings by Era (Peak Industry Period)

```python
df['Decade'] = (df['Entertain_Year'] // 10) * 10

era_stats = df.groupby('Decade').agg({
    'Entertainer': 'count',
    'Total_Career_Earnings_Lakhs': 'mean'
})
#.reset_index()
print(era_stats)
sns.lineplot(data=era_stats, x='Decade', y='Total_Career_Earnings_Lakhs')
plt.title("Average Career Earnings by Era")
plt.show()
```


##  **Problem 10: Earnings by Era (Decade-wise Analysis)**

## **Conclusion:**
Each decade shows varying average earnings, with a noticeable rise in more recent decades.
This reflects industry growth, inflation, and increased monetization avenues over time (e.g., streaming, endorsements).
